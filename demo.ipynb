```python
def get_top_indices(score_list, mode, num_top):
    score_list = pd.Series(score_list)

    if mode == "max":
        sorted_score_list = score_list.sort_values(ascending=False)
    else:
        sorted_score_list = score_list.sort_values(ascending=True)

    result = sorted_score_list.index.to_list()[:num_top]

    return result

score_list = [0.01, 0.4, 1.6, 7.8, 2.4, 1.5, 9.7]
get_top_indices(score_list, "max", 3) # 3, 4, 6

```

```python
from huggingface_hub import hf_hub_download

def get_clip_embed_dim_from_hf_hub(clip_model_name: str) -> int:
    # clip_model_name: "hf-hub:microsoft/BiomedCLIP-PubMedBERT_256-vit_base_patch16_224"
    repo_id = clip_model_name.removeprefix("hf-hub:")
    config_path = hf_hub_download(repo_id=repo_id, filename="open_clip_config.json")

    cfg = kltn_utils.read_json_to_dict(config_path)

    return cfg["model_cfg"]["embed_dim"]


get_clip_embed_dim_from_hf_hub("hf-hub:microsoft/BiomedCLIP-PubMedBERT_256-vit_base_patch16_224")
```
